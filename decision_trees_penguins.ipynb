#"Palmer Penguins"

The model should be built to classify penguins based on their physical features.


Data source: Gorman KB, Williams TD, Fraser WR (2014) Ecological Sexual Dimorphism and Environmental Variability within a Community of Antarctic Penguins (Genus Pygoscelis). PLoS ONE 9(3): e90081. doi:10.1371/journal.pone.0090081
* penguins_size.csv: Simplified data from original penguin data sets. Contains variables:

    * species: penguin species (Chinstrap, Ad√©lie, or Gentoo)
    * culmen_length_mm: culmen length (mm)
    * culmen_depth_mm: culmen depth (mm)
    * flipper_length_mm: flipper length (mm)
    * body_mass_g: body mass (g)
    * island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
    * sex: penguin sex

#Machine Learning

```python
X = pd.get_dummies(df.drop('species',axis=1),drop_first=True)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred = model.predict(X_test)
```

#Model Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
```

```python
confusion_matrix(y_test,base_pred)
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
print(classification_report(y_test,base_pred))
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance'])
```

```python
sns.boxplot(x='species',y='body_mass_g',data=df)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8))
plot_tree(model);
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(model,filled=True,feature_names=X.columns);
```

##Varying different hyperparameters

The default parameters will be changend to a different value like max_depth, criterion and max leaf nodes.

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X.columns);
```

###Max Depth

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

###Max Leaf Nodes

```python
pruned_tree = DecisionTreeClassifier(max_leaf_nodes=3)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

###Criterion

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
